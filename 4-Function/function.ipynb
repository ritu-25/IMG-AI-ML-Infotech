##### Introduction to Functions
Definition:

A function is a block of code that performs a specific task.
Functions help in organizing code, reusing code, and improving readability.



```python
## Syntax

def function_name(parameters):
    """Docstring"""
    
    # Function body 
    return expression
```

```python
## Why functions?
num=24
if num%2==0:
    print("The number is even")
else:
    print("The number is odd")
```

```python
def even_or_odd(num):
    """This function finds evn or odd"""
    if num%2==0:
        print("The number is even")
    else:
        print("The number is odd")
```

```python
## Call This function

even_or_odd(24)
```

```python
## function with multiple parameters

def add(a,b):
    # c = a+b
    # return c
    
    return a+b

result = add(2,4)
print(result)
```

```python
## Default Parameters

def greet(name):
    print(f"Hello {name} welcome to the Paradise")

greet("Ritu")
```

```python
### Variable length Arguments
## Positional And Keywords Arguments

## Positional Arguments
def print_numbers(*Ritu):
    for number in Ritu:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Ritu")
```

```python
### KeywordsArguments 
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
    
```

```python
print_details(name="Ritu",age="22",country="India")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f" Positional Argument :{val}")
        
    print("This is Keyword arguments")
    for key,value in kwargs.items():
        print(f"{key}:{value}")
    
```

```python
print_details(1,2,3,4,name="Ritu",age="22",country="India")
```

```python
### Return Statements

def multiply(a,b):
    return a*b

multiply(3,4)
```

```python
### Return Multiple parameters 

def multiply(a,b):
    return a*b,a

multiply(3,4)
```
